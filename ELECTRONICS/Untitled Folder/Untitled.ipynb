```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

vsg=np.loadtxt('VSG=-2,5.py.csv' , delimiter=',')
print(vsg)

ID=vsg[0:11:1,1]  # slice
#print(ID)

VSD=vsg[0:11:1,0]
#print(VSD);

plt.plot(VSD,ID,'ro-')


#help(plt.xlim)
plt.xlim(.5,-10.5)
plt.ylim(0,-0.5)
plt.title("VSG = -2.5V", size='24',weight='bold')
plt.xlabel("VDS (V)",weight='bold')
plt.ylabel("ID (mA)", weight='bold')
```
