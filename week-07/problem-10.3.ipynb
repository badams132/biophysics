# 10.3 RANDOM WALK WITH RANDOM WAITING TIMES

Use this notebook to store your work for this problem.

## (a)

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rand = default_rng().random
import statistics

beta = 1
N = 400
u = rand(N)
y = -np.log(u)
t = y/beta

time = t.cumsum()


xi = 0.5
steps = 2*(rand(N) < xi)-1
location = steps.cumsum()
#print(location)

steps1 = 2*(rand(N) < xi)-1
location1 = steps1.cumsum()



plt.plot(time, location, 'r-')
plt.plot(time, location1, 'k-')

plt.title("Random walk with random waiting times")
plt.xlabel("time(s)")
plt.ylabel("position(10^-6m)");
```

## (b)

```python
N=400
traj = np.arange(50)
X_T = np.zeros(50)


for i in np.arange(50):
    xi = 0.5
    steps = 2*(rand(N) < xi)-1
    location = steps.cumsum()
    #print(location)
    
    X_T[i]= location[-1]


print(X_T) 
    
    
plt.plot(traj, X_T, 'r-')
var =  statistics.variance(X_T, xbar = X_T.mean())
#print(var)
print ("The sample mean is {}, and the sample Varience is {:.7} for 50 end posistion for 400 seconds".format(X_T.mean(),var))

plt.title("50 Random walks ending positions at 400 seconds")
plt.xlabel("Walks")
plt.ylabel("position(10^-6m)");
```
