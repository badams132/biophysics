## (a) Plot the phase portrait.

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.integrate import odeint

x = np.linspace(-np.pi,np.pi,21)
y_1 = np.linspace(-3,3,21)

X, Y = np.meshgrid(x,y_1)

vX = Y
vY = -X
plt.quiver(X,Y,vX,vY, pivot='mid')
#plt.plot(x,x)

g_L = 1
ci_m = 1/3

def F(theta,t):
    
    return (theta[1], -g_L*np.sin(theta)-ci_m*theta[0])
    
#def F(y,t):
    #dy = [0,0]
    #dy[0] = y[1]
    #dy[1] = -y[0] - g_L*np.sin(.8*t)
    #return dy

t = np.linspace(-np.pi,np.pi,201)
y0 = [-3,1]
y = odeint(F,y0,t)

plt.plot(t,y[:,0], 'r-')
plt.plot(t,y[:,1], 'b--')
```
